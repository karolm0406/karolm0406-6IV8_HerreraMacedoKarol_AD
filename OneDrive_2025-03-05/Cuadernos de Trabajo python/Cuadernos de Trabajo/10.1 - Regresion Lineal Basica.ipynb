## Relaciones lineales

Armemos una relación entre Celsius y Farenheit

Armemos una comprensión de lista con esta función para todas las temeperaturas del 0 al 100 de 10 en 10

```python
celsius = [temp for temp in range(0, 101, 10)]  
fahrenheit = [1.8 * temp + 32 for temp in celsius]  
```

```python
print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

Importa Pandas

```python
import pandas as pd
```

Creamos un DataFrame con las listas

```python
df = pd.DataFrame({'Celsius': celsius, 'Fahrenheit': fahrenheit})
```

```python
print(df)
```

Arma una gráfica que muestre esta relación

```python
import matplotlib.pyplot as plt

plt.scatter(df['Celsius'], df['Fahrenheit'], color='blue', label='Datos reales') 
plt.xlabel('Temperatura en Celsius (°C)')
plt.ylabel('Temperatura en Fahrenheit (°F)')
plt.title('Relación entre Celsius y Fahrenheit')
plt.legend()
plt.show()
```

Ponle etiquetas

y = mx + b

```python
x = df['Celsius']
y = [1.8 * temp + 32 for temp in x] 

plt.scatter(df['Celsius'], df['Fahrenheit'], color='blue', label='Datos reales')  
plt.plot(x, y, color='red', label='Regresión Lineal: F = 1.8C + 32') 

plt.xlabel('Temperatura en Celsius (°C)')
plt.ylabel('Temperatura en Fahrenheit (°F)')
plt.title('Relación Lineal: Celsius vs Fahrenheit')
plt.legend()
plt.show()
```
